# **SPLITTING**

```python
x = df.drop("stroke",axis=1)
y = df["stroke"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
y_train.value_counts()
```

# **SMOTE**

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
```

```python
x_res,y_res = sm.fit_resample(x_train,y_train)
```

```python
y_res.value_counts()
```

# **STANDARDIZATION**

```python
from sklearn.preprocessing import StandardScaler
sd = StandardScaler()
x_train = sd.fit_transform(x_train)
x_test = sd.fit_transform(x_test)
x_res = sd.fit_transform(x_res)
```

# **DECISION TREE MODEL**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
class_report = classification_report(y_test, y_pred)
print(class_report)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix
```

```python
train_score = dt.score(x_train,y_train)
train_score
```

```python
test_score = dt.score(x_test,y_test)
test_score
```

# **DECISION TREE WITH SOMTE DATA**

```python
accuracy_sm = accuracy_score(y_test, y_pred_smote)
accuracy_sm
```

```python
class_report_sm = classification_report(y_test, y_pred_smote)
print(class_report_sm)
```

```python
test_score_sm = dt_smote.score(x_test,y_test)
test_score_sm
```
